```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件

# 忽略警告訊息
%matplotlib inline
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')

# 設定 data_path
dir_data = '/home/osboxes/Documents/data_set/'
```

```python
# 讀取檔案
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
data_house = app_train[["HOUSETYPE_MODE", "TARGET"]]
house = data_house["HOUSETYPE_MODE"].dropna(axis=0, how="any")
huse_group_sorted = np.sort(house.unique())
print(data_house)
print(house)
print(huse_group_sorted)
#print(np.sort(data_house["HOUSETYPE_MODE"].unique()))
print(data_house["HOUSETYPE_MODE"].value_counts())
```

```python
# 使用不同的 HOUSETYPE_MODE 類別繪製圖形, 並使用 subplot 排版
"""
Your Code Here
"""
unique_house_type = np.sort(house.unique())

nrows = len(unique_house_type)
ncols = nrows // 2

plt.figure(figsize=(10,30))
for i in range(len(unique_house_type)):
    plt.subplot(nrows, ncols+1, i+1)
    """
    Your Code Here
    """
    app_train.loc[ (data_house['HOUSETYPE_MODE'] == unique_house_type[i]) &\
                  (data_house["TARGET"] == 0), "HOUSETYPE_MODE"].hist()
    app_train.loc[ (data_house['HOUSETYPE_MODE'] == unique_house_type[i]) &\
                  (data_house["TARGET"] == 1), "HOUSETYPE_MODE"].hist()
    
    
    plt.title(str(unique_house_type[i]))
plt.show()    
```
