```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score
```

```python
boston = datasets.load_boston()
print(boston.data)
data = boston.data[:, np.newaxis, 2]
print(data)
```

```python
MMEncoder = MinMaxScaler()
data = MMEncoder.fit_transform(data)
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, boston.target, test_size=0.1, random_state=4)
```

```python
print(y_test)
```

```python
LR = linear_model.LinearRegression()
LR.fit(x_train, y_train)
```

```python
y_pred = LR.predict(x_test)
```

```python
print('Coefficients: ', LR.coef_)
print("Mean squared error: %.2f"
      % mean_squared_error(y_test, y_pred))
```

```python
# 畫出回歸模型與實際資料的分佈
plt.scatter(x_test, y_test,  color='black')
plt.plot(x_test, y_pred, color='blue', linewidth=3)
plt.show()
```

```python
wine = datasets.load_wine()
print(wine.data)
data = wine.data[:, np.newaxis, 2]
print(data)
```

```python
MMEncoder = MinMaxScaler()
data = MMEncoder.fit_transform(data)
x_train, x_test, y_train, y_test = train_test_split(data, wine.target, test_size=0.1, random_state=4)
```

```python
Logistic_Regression = linear_model.LogisticRegression()
Logistic_Regression.fit(x_train, y_train)
```

```python
y_pred = Logistic_Regression.predict(x_test)
acc = accuracy_score(y_test, y_pred)
print("Accuracy: ", acc)
```
